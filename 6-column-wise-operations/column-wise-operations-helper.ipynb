```python
import pandas as pd
import seaborn as sns

tips = sns.load_dataset('tips')
tips.head()
```

```python
tips['tip'] / tips['total_bill']
tips.head()
```

```python
tips['column'] = (tips['tip'] / tips['total_bill']).round(2)
tips.head()
```

```python
tips.rename(columns = {'column': 'tip_pct'}, inplace=True)
tips.columns
```

```python
tips['day'] = tips['day'].str.upper()
tips.head()
```

```python
tips['running_total_bill'] = tips['total_bill'].cumsum()
tips.head()
```

```python
tips = tips.drop('running_total_bill', axis='columns')
```

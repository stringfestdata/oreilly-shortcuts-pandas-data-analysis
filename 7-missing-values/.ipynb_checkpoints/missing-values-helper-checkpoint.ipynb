```python
import pandas as pd
import seaborn as sns

penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
penguins.isna()
```

```python
penguins.isna().sum()
```

```python
penguins.isna().sum()/len(penguins)
```

```python
penguins.dropna(inplace=True)
penguins.isna().sum()/len(penguins)
```

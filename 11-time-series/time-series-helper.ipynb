```python
#! pip install vega_datasets
import pandas as pd
from vega_datasets import local_data
import seaborn as sns
import matplotlib.pyplot as plt

seattle = local_data.seattle_weather()
seattle.head()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x='date', y='precipitation', data=seattle)
```

```python
seattle_ts = seattle.set_index('date')
seattle_ts.head()
```

```python
monthly_precip = seattle_ts.resample('ME').sum()['precipitation']
monthly_precip.head()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x=monthly_precip.index, y=monthly_precip.values)
```

```python
seattle['precip_rolling_30'] = seattle['precipitation'].rolling(30).sum()
seattle.head(30)
```

```python
sns.lineplot(data=seattle, x='date', y='precip_rolling_30')
```

```python
seattle['precip_lead'] = seattle['precipitation'].shift(-1)
seattle['precip_lag'] = seattle['precipitation'].shift(1)
```

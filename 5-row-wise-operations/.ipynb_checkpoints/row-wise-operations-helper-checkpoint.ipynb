```python
import pandas as pd
import seaborn as sns

tips = sns.load_dataset('tips')
tips.head()
```

## Sorting

```python
# Sort by total bill
tips.sort_values(by=['total_bill'])
```

```python
# Total bill, descending
tips.sort_values(by=['total_bill'], ascending=False)
```

```python
tips
```

```python
tips_sorted = tips.sort_values(by=['total_bill'], ascending=False)
tips_sorted.head()
```

```python
# Time asecnding, total bill descending
tips.sort_values(by=['time', 'total_bill'], ascending=[True, False])
```

## Filtering

```python
# Filtering tip amount greater than 5
tips_over_5 = tips['tip'] > 5
tips_over_5
```

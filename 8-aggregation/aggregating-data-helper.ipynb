```python
import pandas as pd
import seaborn as sns

tips = sns.load_dataset('tips')
tips.head()
```

```python
tips_grouped = tips.groupby('day', observed=True)
tips_grouped
```

```python
tips_grouped[['total_bill']].mean()
```

```python
tips.groupby('time', observed=True)[['tip']].max()
```

```python
tips.groupby(['day', 'time'], observed=True)[['tip']].max()
```

```python
tips.groupby(['day', 'time'], observed=True)[['tip']].agg(['mean', 'sum'])
```
